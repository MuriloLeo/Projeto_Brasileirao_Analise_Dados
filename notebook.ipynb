```python
df_brasileirao[df_brasileirao.isna().any(axis=1)]
display(df_brasileirao)

# Altera tipo das colunas
df_brasileirao['Empate'] = df_brasileirao['Empate'].astype(bool)
df_brasileirao['Vencedor'] = df_brasileirao['Vencedor'].astype(str)
df_brasileirao['Derrotado'] = df_brasileirao['Derrotado'].astype(str)

df_brasileirao['Pontuacao_Casa'] = df_brasileirao['Pontuacao_Casa'].astype(int)
df_brasileirao['Pontuacao_Visitante'] = df_brasileirao['Pontuacao_Visitante'].astype(int)

df_brasileirao['Num_Vitorias_Casa'] = df_brasileirao['Num_Vitorias_Casa'].astype(int)
df_brasileirao['Num_Derrota_Casa'] = df_brasileirao['Num_Derrota_Casa'].astype(int)
df_brasileirao['Num_Empate_Casa'] = df_brasileirao['Num_Empate_Casa'].astype(int)

df_brasileirao['Num_Vitorias_Visitante'] = df_brasileirao['Num_Vitorias_Visitante'].astype(int)
df_brasileirao['Num_Derrota_Visitante'] = df_brasileirao['Num_Derrota_Visitante'].astype(int)
df_brasileirao['Num_Empate_Visitante'] = df_brasileirao['Num_Empate_Visitante'].astype(int)

df_brasileirao['Temporada'] = df_brasileirao['Temporada'].astype(int)
df_brasileirao['Rodada'] = df_brasileirao['Rodada'].astype(int)
```

```python
df_brasileirao['Saldo_Mandante'] = df_brasileirao['Gols_Mandante'] - df_brasileirao['Gols_Visitante']
df_brasileirao['Saldo_Visitante'] = df_brasileirao['Gols_Visitante'] - df_brasileirao['Gols_Mandante']

# 2️⃣ Criando dataframe "long" com todos os times
mandante = df_brasileirao[['Temporada','Time_Casa','Saldo_Mandante']].rename(
    columns={'Time_Casa':'Time','Saldo_Mandante':'Saldo_Gols'})
visitante = df_brasileirao[['Temporada','Time_Visitante','Saldo_Visitante']].rename(
    columns={'Time_Visitante':'Time','Saldo_Visitante':'Saldo_Gols'})

todos = pd.concat([mandante, visitante])

# 3️⃣ Somando saldo de gols por time e temporada
saldo_por_time = todos.groupby(['Temporada','Time'])['Saldo_Gols'].sum().reset_index()
```

```python
tabela_pontos = pd.concat([
    df_brasileirao[['Temporada', 'Time_Casa', 'Pontuacao_Casa', 'Num_Vitorias_Casa', 'Num_Derrota_Casa', 'Num_Empate_Casa']].rename(
        columns={'Time_Casa':'Time', 'Pontuacao_Casa':'Pontos', 'Num_Vitorias_Casa': 'Vitorias', 'Num_Derrota_Casa': 'Derrotas', 'Num_Empate_Casa': 'Empates'}
    ),
    df_brasileirao[['Temporada', 'Time_Visitante', 'Pontuacao_Visitante', 'Num_Vitorias_Visitante', 'Num_Derrota_Visitante', 'Num_Empate_Visitante']].rename(
        columns={'Time_Visitante':'Time', 'Pontuacao_Visitante':'Pontos', 'Num_Vitorias_Visitante': 'Vitorias', 'Num_Derrota_Visitante': 'Derrotas', 'Num_Empate_Visitante': 'Empates'}
    )
])

tabela_classificacao = tabela_pontos.groupby(['Temporada', 'Time'], as_index=False).agg({
    'Pontos': 'sum',
    'Vitorias': 'sum',
    'Empates': 'sum',
    'Derrotas': 'sum'
}).sort_values(['Temporada', 'Pontos'], ascending=[True, False])

tabela_classificacao.loc[tabela_classificacao['Temporada'] == 2022]


```

```python
df_classificacao_agrupada = tabela_classificacao.groupby(['Temporada', 'Time'], as_index=False).agg({
    'Pontos': 'sum',
    'Vitorias': 'sum',
    'Empates': 'sum',
    'Derrotas': 'sum'
}).sort_values('Pontos', ascending=[False])

df_final = df_classificacao_agrupada.merge(
    saldo_por_time,
    on=['Temporada','Time'],
    how='left'
)
```

```python
df_final = df_final.sort_values(['Pontos'], ascending=[False])

df_final.loc[df_final['Temporada'] == 2022]
```

```python
# Para mandantes
df_mandante = df_final.rename(columns={
    'Time': 'Time_Casa',
    'Pontos': 'Total_Pontos_Casa',
    'Vitorias': 'Total_Vitorias_Casa',
    'Empates': 'Total_Empates_Casa',
    'Derrotas': 'Total_Derrotas_Casa',
    'Saldo_Gols': 'Total_Saldo_Casa'
})

# Para visitantes
df_visitante = df_final.rename(columns={
    'Time': 'Time_Visitante',
    'Pontos': 'Total_Pontos_Visitante',
    'Vitorias': 'Total_Vitorias_Visitante',
    'Empates': 'Total_Empates_Visitante',
    'Derrotas': 'Total_Derrotas_Visitante',
    'Saldo_Gols': 'Total_Saldo_Visitante'
})

# Junta os dados do time mandante
df_dashboard = df_brasileirao.merge(df_mandante, on=['Temporada', 'Time_Casa'], how='left')

# Junta os dados do time visitante
df_dashboard = df_dashboard.merge(df_visitante, on=['Temporada', 'Time_Visitante'], how='left')
```

```python
df_final.to_csv('classificacao_final_brasileirao.csv', index=False)
df_dashboard.to_csv('dashboard_brasileirao.csv', index=False)
```
